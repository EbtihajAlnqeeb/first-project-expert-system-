```python

from experta import *

class Product(Fact):

  #attribute that we will test its results

  toothpaste =Field(str)
  start=Field(str)
  adult =Field(str)
  sensitive =Field(str)
  whitening =Field(str)
  diabetes =Field(str)
  natural =Field(str)
  gingivitis =Field(str)
  pain= Field(str)
  dryMouth= Field(str)
  toothRemoved= Field(str)
  nerveInflammation=Field(str)
  
  pass

class Pharamcy(KnowledgeEngine):
 
#Rule 1:
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('yes')), Product(whitening=('yes')),Product(sensitive=('no'))))
  
  def select_whitening(self):

   print("\n  The avalibe products for your situation are: \n 1-Oral-B 3D white Luxe Perfection. \n 2-Sensodyne Extra Whiting Teeth.")

   print(" for best result : Brush teeth thoroughly for at least 1 minute twice a day ,not more than 3 times a day! \n\n")


#Rule2
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('yes')), Product(whitening=('yes')),Product(sensitive=('yes'))))
  
  def select_sensitiveWhitening(self):

    print("\n The avalibe products for your situation are: \n 1-Cogate Sensitive whiting Toothpast. \n 2-sensodyne Sensitive Teeth.")

    print (" for best result : Brush teeth thoroughly for at least 1 minute twice a day ,not more than 3 times a day! \n\n ")

#Rule3
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('yes')), Product(whitening=('no')), Product(gingivitis=('yes'))))
  
  def select_gingivitis(self):

    print("\nThe avalibe products for your situation are: \n 1-parodontax Extra Fresh  Toothpastes for Gingiviti.")

    print(" Use twice daily. Rinse with 10ml for one minute then spit out.")

#Rule4
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('yes')), Product(whitening=('no')), Product(gingivitis=('no')),Product(diabetes =('yes'))))
  
  def select_diabetes(self):

    print("\n The avalibe products for your situation are: \n 1-colgate Toothpaste For Diabetics that has no sugar")

#Rule5
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('yes')), Product(whitening=('no')),
            Product(gingivitis=('no')),Product(diabetes =('no')),Product(natural =('yes'))))
  
  def select_natural(self):

    print("\n The avalibe products for your situation are: \n 1-USDA organic Mint Aloe Neem Toothpaste")


#Rule 6
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('yes')), Product(whitening=('no')),
            Product(gingivitis=('no')),Product(diabetes =('no')),Product(natural =('no'))))
  
  def select_normal(self):

    print("\n The avalibe products for your situation are: \n 1-signal Toothpase \n 2-Sensodyne Toothpast. \n 3-Oral-B Toothpaste.")

#Rule7
  @Rule(AND(Product(start=('yes')),Product(adult=('no')),Product(toothpaste=('yes'))))
  
  def select_child(self):

    print("\n The avalibe products for your situation are: \n 1-Crest toothpaste for Kids ( has less Fluoride ). \n 2-AloeDent for kids(Fluoride Free).")   

#Rule8
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('no')),Product(dryMouth=('yes'))))
  
  def select_dry(self):

    print("\n The avalibe products for your situation are: \n 1-Oral-B Dry Moth \n 2-Oral Rinse  Mouthwash ")
    

#Rule9
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('no')),Product(dryMouth=('no')),Product(pain=('yes')),Product(deacy=('yes'))))
  
  def select_pain(self):

    print("\n The avalibe products for your situation is: \n 1-Acetaminophen 225 mg ")

    print(" Empty the powder from each capsule into 1 teaspoonful (5 mL) of water or other liquid. Drink the medicine along with the liquid")


#Rule10
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('no')),Product(dryMouth=('no')),Product(pain=('yes')),Product(deacy=('no')),
            Product(toothRemoved=('yes'))))
  def select_toothRemoved(self):
    print("\n The avalibe products for your situation is: \n 1-I-Brofen 200 mg")
    print("The usual dose for adults is one or two 200mg tablets or capsules 3 times a day.")


#Rule11
  @Rule(AND(Product(start=('yes')),Product(adult=('yes')),Product(toothpaste=('no')),Product(dryMouth=('no')),Product(pain=('yes')),Product(deacy=('no')),
            Product(toothRemoved=('no')),Product(nerveInflammation=('yes'))))
  def select_nerveInflammation(self):
    print("\n The avalibe products for your situation is: \n Rofenac 50 mg")
    print("Take with or right after a meal 2 or 3 times a day.")




engine =Pharamcy()
#prepare the engine for excution:

engine.reset()

Product_start =' ';
Product_adult=' ';
Product_toothpaste=' ';
Product_whitening=' ';
Product_sensitive=' ';

Product_diabetes=' ';
Product_gingivitis=' ';
Product_natural=' ';

Product_pain=' ';
Product_deacy=' ';
Product_dryMouth=' ';
Product_toothRemoved=' ';
Product_nerveInflammation=' ';



#welcoming massage and showing our products:

print("-------------------------------------------------------------------")
print("------------------welcome to youre smile system--------------------")
print("-------------------------------------------------------------------")
print("our system will help you select good product for your oral health ")




#The start of pharmacy service and taking input from the user:


Product_start=input(" \n Do you want to start the test? (Yes/NO) ").lower()
if Product_start == 'yes':
 print("start")
 Product_adult=input(" \n Are you older than 12 years? (Yes/NO) ").lower()
 if Product_adult =='yes':
  Product_toothpaste=input("\n Do you want to buy toothpaste? (Yes/NO) ").lower()
  if Product_toothpaste =='yes':
   Product_whitening=input("\n Do you want whitening-toothpaste type? (Yes/NO) ").lower()
   if Product_whitening=='yes':
      Product_sensitive=input("\n Do you suffer from tooth sensitivity? (Yes/NO) ").lower()
   if Product_whitening=='no':
    Product_gingivitis=input("\n Do you suffer from gingivitis? (Yes/NO)  ").lower()
    if Product_gingivitis=='no':
      Product_diabetes=input("\n Are you a diabetic person? (Yes/NO) ").lower()
      if Product_diabetes=='no':
       Product_natural=input("\n Do you prefer to use natural products for your teeth? (Yes/NO) ").lower()
  if Product_toothpaste=='no':
    Product_dryMouth=input("\n Do you suffer from dry-mouth? (YES/NO) \n ").lower()
    if Product_dryMouth=='no':
      Product_pain=input("\n Do you have a pain problem in your maouth? (YES/NO)\n ").lower()
      if Product_pain=='yes':
        Product_deacy=input("\n Do you have deacayed teeth? (YES/NO)\t").lower()
        if Product_deacy=='no':
          Product_toothRemoved=input("\n Do you have tooth removed? (YES/NO)\t").lower()
          if Product_toothRemoved=='no':
           Product_nerveInflammation=input("\n Do you have nerve-Inflammation? (YES/NO)\t").lower()
      
 if Product_adult=='no':
    Product_toothpaste=input("\n Do you want to buy toothpaste for your child? (Yes/NO)\t").lower()

 #print("Thank you for visting our system ,we wish for you a happy healthy smile!")

 


else:
 print("\n maybe next time,goodbye!\n ")
 








engine.declare(Product(start=Product_start),Product (adult=Product_adult),Product(toothpaste=Product_toothpaste),Product(whitening=Product_whitening),
               Product(sensitive=Product_sensitive),Product(gingivitis=Product_gingivitis),Product(diabetes=Product_diabetes),Product(natural=Product_natural),
               Product(pain=Product_pain),Product(deacy=Product_deacy),Product(dryMouth=Product_dryMouth),
               Product(toothRemoved=Product_toothRemoved),Product(nerveInflammation=Product_nerveInflammation)
  )



engine.run()

print(" \n \n Thank you for visting our system ,we wish for you a happy healthy smile! \n\n")

```
